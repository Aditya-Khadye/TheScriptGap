```python
import pandas as pd
import plotly.express as px
```

### Exposure Graphs

```python
# EXPOSURE CSV
exp = pd.read_csv("data/exp.csv")
```

```python
exp
```

```python
fig = px.bar(
    exp,
    x="count",
    y="script",
    title="Script Exposure",
    text="count"
)

fig.update_traces(textposition="outside")

fig.show()
```

```python
fig = px.bar(
    exp[exp["script"] != "Latin"],
    x="count",
    y="script",
    title="Script Exposure",
    text="count"
)

fig.update_traces(textposition="outside")

fig.show()
```

### Script Support (Just Global)

```python
# EXPOSURE CSV
sup = pd.read_csv("data/sup.csv")
```

```python
fig = px.bar(
    sup,
    x="distinct_font_count",
    y="script",
    title="Script Support",
    text="distinct_font_count"
)

fig.update_traces(textposition="outside")

fig.show()
```

```python
fig = px.bar(
    sup[sup["script"] != "Latin"],
    x="distinct_font_count",
    y="script",
    title="Script Support",
    text="distinct_font_count"
)

fig.update_traces(textposition="outside")

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

merged = pd.merge(exp, sup, on="script", how="outer").fillna(0)
merged = merged.sort_values("count", ascending=False)

fig = go.Figure()

fig.add_trace(go.Bar(
    name="Font Request Count",
    x=merged["script"],
    y=merged["count"],
    marker_color="#4C78A8",
    yaxis="y1",
    offsetgroup=1,
))

fig.add_trace(go.Bar(
    name="Distinct Fonts",
    x=merged["script"],
    y=merged["distinct_font_count"],
    marker_color="#F58518",
    yaxis="y2",
    offsetgroup=2,
))

fig.update_layout(
    title="A Script's Number of Web Requests vs Distinct Font Count (Log Scale)",
    xaxis=dict(title="Script", tickangle=-35),
    yaxis=dict(
        title="Font Request Count",
        title_font=dict(color="#4C78A8"),
        tickfont=dict(color="#4C78A8"),
        type="log",  # <-- add this
    ),
    yaxis2=dict(
        title="Distinct Font Count",
        title_font=dict(color="#F58518"),
        tickfont=dict(color="#F58518"),
        overlaying="y",
        side="right",
        type="log",  # <-- add this
    ),
    barmode="group",
    bargap=0.2,
    bargroupgap=0.05,
    legend=dict(x=0.75, y=0.97),
    template="plotly_white",
    width=700,
    height=700
)

fig.show()
```
