```python
import pandas as pd
import plotly.express as px
```

### Exposure Graphs

```python
# EXPOSURE CSV
exp = pd.read_csv("data/exp.csv")
```

```python
exp
```

```python
fig = px.bar(
    exp,
    x="count",
    y="script",
    title="Script Exposure",
    text="count"
)

fig.update_traces(textposition="outside")

fig.show()
```

```python
fig = px.bar(
    exp[exp["script"] != "Latin"],
    x="count",
    y="script",
    title="Script Exposure",
    text="count"
)

fig.update_traces(textposition="outside")

fig.show()
```
